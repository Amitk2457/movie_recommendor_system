#**Content** **based** **filtering**





```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/movies_data/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/movies_data/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on = 'title')
```

```python
movies.head(1)
```

```python
movies.shape
```

## Data Preprocessing

```python
movies.info()
```

```python
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
movies.head()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
#Extracting Genre
import ast
def extract(obj):
  lis = []
  for i in ast.literal_eval(obj):
    lis.append(i['name'])

  return lis
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
def extract3(obj):
  count = 0
  lis = []
  for i in ast.literal_eval(obj):
    if count < 3 :
        lis.append(i['name'])
        count = count +1

  return lis
```

```python
movies['cast'] = movies['cast'].apply(extract3)
```

```python
def extractDirector(obj):
  lis = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director' :
        lis.append(i['name'])
        break

  return lis
```

```python
movies['crew'] = movies['crew'].apply(extractDirector)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] +movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
nltk.download('punkt')
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i) )

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)

tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

tfidf_matrix = tfidf_vectorizer.fit_transform(new_df['tags'])

vector = tfidf_matrix.toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
tfidf_vectorizer.get_feature_names_out()
```

##Cosine Similarity for Content Based Filtering

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity.shape
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  dist = similarity[movie_index]
  temp = sorted(list(enumerate(dist)),reverse=True,key = lambda x:x[1])[1:6]

  for i in temp:
    Z = new_df.iloc[i[0]].title
    print(Z)
```

```python
recommend('Batman Begins')
```
